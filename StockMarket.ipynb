```python
#import libraries
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
#defining a list of assets that will be analized
assets=["AAPL","MSFT","TSLA","MMM"]
```

```python
#pulling data for the assest by using yfinace and providing the period frame
df=yf.download(assets,period="1y")
```

```python
#print df
df
```

```python
#for this analysis Adj close is only column needed
df['Adj Close']
```

```python
prices=df['Adj Close']
```

```python
prices
```

```python
#daily returns of apple for a given day
176.880646/178.263306-1
```

```python
#getting the daily change in each assets
prices.pct_change(fill_method=None)
```

```python
#finidng how apple was performing from a certain date to another
(0.022816+1)*(-0.009722 +1)
```

```python
#storing returns in this variable
returns=prices.pct_change(fill_method=None)
returns
```

```python
#cumulative return in all rows,also know how much stock market was rising from 2023-2024
(returns+1).cumprod()-1
cum_ret=(returns+1).cumprod()-1
cum_ret
```

```python
cum_ret.plot(figsize=(10,6))
```

```python
#volatility-mesasurement usually calculated using standard deviation(showing how much the assest is wining around it own mean
returns.std()
```

```python
#Visual comparison of multiple assets using subplots
fig,ax =plt.subplots(2,2, figsize=(16,5))
plt.subplots_adjust(bottom=0.1, top=0.9, wspace=0.3, hspace=0.4)

Counter=0
for i in range(2):
    for j in range(2):
        ax[i,j].plot(returns[returns.columns[Counter]])
        ax[i,j].set_title(returns.columns[Counter])
        Counter +=1
    
```

```python
#Visual comparison of multiple assets for Return distributions
fig,ax =plt.subplots(2,2, figsize=(16,5))
plt.subplots_adjust(bottom=0.1, top=0.9, wspace=0.3, hspace=0.4)

Counter=0
for i in range(2):
    for j in range(2):
        ax[i,j].hist(returns[returns.columns[Counter]],bins=50)
        ax[i,j].set_title(returns.columns[Counter])
        Counter +=1
    
```
